```python
import numpy as np
from matplotlib import pyplot as plt
np.random.seed(70070543)
```

```python
X = np.random.randn(50)
Y1 = 0.8*X + 0.5*np.random.randn(50)
```

```python
plt.plot(X, Y1, 'o')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.title('A', size=30)
plt.xlabel('X', size=26)
plt.ylabel('Y', size=26)
plt.tight_layout()
plt.savefig('../figs/fig1.png')
```

```python
Y2 =  0.3*np.random.randn(50)
plt.plot(X, Y2, 'o')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.title('B', size=30)
plt.xlabel('X', size=26)
plt.ylabel('Y', size=26)
plt.tight_layout()
plt.savefig('../figs/fig2.png')
```

```python
Y3 =  np.random.randn(50)
plt.plot(X, Y3, 'o')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.title('C', size=30)
plt.xlabel('X', size=26)
plt.ylabel('Y', size=26)
plt.tight_layout()
plt.savefig('../figs/fig3.png')
```

```python
Y4 = 0.5*(np.random.randn(50) - X*X) + 2
plt.plot(X, Y4, 'o')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.title('D', size=30)
plt.xlabel('X', size=26)
plt.ylabel('Y', size=26)
plt.tight_layout()
plt.savefig('../figs/fig4.png')
```

```python
X2 = X / 2 + 2
X2[X<-1.5] = -2.5 + 0.2*np.random.randn(sum(X<-1.5)) 
Y5 = np.random.randn(50)
Y5[X<-1.5] = -2.5 + 0.2*np.random.randn(sum(X<-1.5)) 
plt.plot(X2, Y5, 'o')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.title('E', size=30)
plt.xlabel('X', size=26)
plt.ylabel('Y', size=26)
plt.tight_layout()
plt.savefig('../figs/fig5.png')
```

```python
Y5 = -3*X + 0.5*np.random.randn(50) + 6
Y5[X<1] = -3*X[X<1] + 0.5*np.random.randn(sum(X<1))
Y5[X<-1] = -3*X[X<-1] + 0.5*np.random.randn(sum(X<-1)) - 6.5
plt.plot(X, Y5, 'o')
#plt.xlim(-3, 3)
#plt.ylim(-3, 3)
plt.xlabel('X', size=26)
plt.ylabel('Y', size=26)
plt.title('F', size=30)
plt.tight_layout()
plt.savefig('../figs/fig6.png')
```
